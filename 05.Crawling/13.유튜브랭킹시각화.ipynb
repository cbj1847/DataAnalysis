```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('data/20230425_YouTube랭킹.csv')
df.head(3)
```

- 구독자수 Top 20 - barh 차트로

```python
df1 = df.sort_values(by='구독자수', ascending=False)
df1.head(20)
```

```python
plt.barh(df1.채널명[:20] ,df1.구독자수[:20])
plt.show()
```

```python
sns.barplot(y=df1.채널명[:20], x=df1.구독자수[:20], data=df1)
plt.title('구독자수에 따른 순위');
```

- 조회수 Top 20

```python
df2 = df.sort_values(by='조회수', ascending=False)
df2.head()
```

```python
sns.barplot(y=df2.채널명[:20], x=df2.조회수[:20])
plt.title('조회수에 따른 순위');
```

- 비디오수 Top 20

```python
df3 = df.sort_values(by='비디오', ascending=False)
df3.head()
```

```python
sns.barplot(y=df3.채널명[:20], x=df3.비디오[:20], data=df3)
plt.title('영상수에 따른 순위');
```

- 카테고리별 채널수 Top 10

```python
cat_cnt = df.groupby('카테고리').count()
cat_cnt
```

```python
cat_cnt = cat_cnt.sort_values(by='채널명',ascending=False)
cat_cnt.head()
```

```python
sns.barplot(y=cat_cnt.index[:10], x=cat_cnt.채널명[:10], data=cat_cnt)
plt.title('카테고리별 채널수에 따른 순위');
```

- 카테고리별 구독자수 Top 10

```python
cat_sub = df.groupby('카테고리').sum()
cat_sub
```

```python
cat_sub = cat_sub.sort_values(by='구독자수', ascending=False)
cat_sub
```

```python
sns.barplot(y=cat_sub.index[:10], x=cat_sub.구독자수[:10], data=cat_sub)
plt.title('카테고리별 구독자수에 따른 순위');
```
