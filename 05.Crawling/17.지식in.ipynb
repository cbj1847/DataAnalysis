```python
url = 'https://kin.naver.com/search/list.naver?query=%EC%96%B4%EB%B2%84%EC%9D%B4%EB%82%A0+%EC%84%A0%EB%AC%BC'
driver = webdriver.Chrome('C:/Users/YONSAI/Desktop/chromedriver_win32/chromedriver.exe')
driver.get(url)
lis = driver.find_elements(By.XPATH, '//*[@id="s_content"]/div[3]/ul/li')
```

```python
# 한 페이지 내의 글 클릭

text_list = []
for li in lis:  
    href = li.find_element(By.CSS_SELECTOR, 'dt > a').get_attribute('href')
    driver.get(href)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    texts = soup.select('.se-module.se-module-text')
    for text in texts:
        text_list.append(text.get_text())
```

```python
driver.close()
```
