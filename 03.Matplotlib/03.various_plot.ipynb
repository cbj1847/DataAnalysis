#### 2. Various Plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

1. Bar Chart

```python
y, x = [2, 3, 1], [0, 1, 2]
plt.title('Bar Chart')
plt.bar(x, y)
plt.xticks(x, ['가','나','다']), plt.yticks(sorted(y))
plt.xlabel('가나다'), plt.ylabel('빈도 수')
plt.show()
```

```python
s = pd.Series(np.random.rand(10), index=list('abcdefghij'))
plt.bar(s.index, s.values)
plt.title('Bar chart'), plt.grid()
plt.yticks([0, 0.2, 0.4, 0.6, 0.8]), plt.xticks([])
plt.show()
```

```python
s.plot(kind='bar')
plt.show()
```

```python
df = pd.DataFrame(np.random.rand(20).reshape(10, 2), index=list('abcdefghij'), columns=['M', 'N'])
plt.bar(df.index, df.M.values)
plt.title('Bar chart'), plt.grid()
plt.show()
```

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic['class'].value_counts()
```

```python
s = titanic['class'].value_counts()
plt.barh(s.index, s.values)
plt.title('타이타닉호의 선실별 승객수')
plt.show()
```

#### 2.2 Pie chart

```python
ratio = [10, 15, 20, 25]        # 합이 100이 아니여도 됨
labels = ['감', '귤', '배', '사과']
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.pie(ratio, labels=labels, autopct='%.1f%%', colors=colors, explode=explode
        , startangle=90, shadow=True)
plt.title('파이 차트')
plt.show()
```

#### 4. 산점도(Scatter plot)

```python
np.random.seed(2023)
X, Y = np.random.randn(100), np.random.normal(2, 2, 100)
plt.scatter(X, Y)
plt.title('Scatter plot')
plt.grid()
plt.show()
```

```python
colors = np.random.rand(100)
plt.scatter(X, Y, c=colors, alpha=0.5)      # c:색상, alpha:투명도
plt.title('Colored Scatter plot')
plt.grid()
plt.show()
```

```python
s = 500 * np.random.rand(100)
plt.scatter(X, Y, c=colors, alpha=0.5, s=s)     #s:알크기
plt.title('Bubble chart')
plt.grid()
plt.show()
```

#### 6. Boxplot

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
ssl = iris[iris.species == 'setosa'].sepal_length
ssw = iris[iris.species == 'setosa'].sepal_width

plt.boxplot((ssl, ssw), labels=['sepal length', 'sepal width'])
plt.title('Setosa 품종의 꽃받침 길이/너비')
plt.grid(color='mediumpurple')
plt.show()
```
