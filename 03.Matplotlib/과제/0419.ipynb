1. titanic
    - embark_town, sex => 생존율
    - 나이 displot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
titanic = sns.load_dataset('titanic')
pct = (titanic.pivot_table('survived', 'embark_town', 'sex') * 100).round(2)
pct
```

```python
pct.plot(kind='bar')
plt.show()
```

```python
fig = sns.displot(x='age', data=titanic, kde=True);
```

2. CCTV
    - 자치구 법정 순, 자치구 이름, CCTV 댓수

```python
df = pd.read_csv('../data/서울시 자치구 년도별 CCTV 설치 현황_221231기준.csv', encoding='euc-kr', thousands=',')
df.drop(index=0, inplace=True)
df.sort_values(by=['구분'], inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
df['2020이후'] = df.iloc[:,-3:].sum(axis=1)
df['2020이전'] = df.iloc[:, 1] - df['2020이후']
df
```

```python
fig = plt.figure(figsize=(16,6))
plt.plot(df.구분, df.iloc[:,-2], label='2020이후', c='b')
plt.plot(df.구분, df.iloc[:,-1], label='2020이전', c='r')
plt.legend()
plt.title('CCTV 총 댓수'), plt.grid(color='lightgray')
plt.show()
```
