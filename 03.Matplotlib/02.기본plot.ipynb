#### 1. Matplotlib 소개

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

#### 1. 라인플롯

```python
plt.plot([0,1,4,9])         # X가 생략되면 y의 갯수만큼 zero indexing
plt.title('라인 플롯')
plt.grid()
plt.show()          # 생략가능
```

- 1.1 스타일 지정

```python
plt.plot([0,1,4,9], 'rs--')         # color, marker, linestyle
plt.title("'rs--' 스타일의 플롯")
plt.grid()
plt.show()
```

```python
plt.plot([0,1,4,9], 'rs--')
plt.plot([9,4,1,0], 'g^:')
plt.title("'g&:' 스타일의 플롯")
plt.grid()
plt.show()
```

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 'bo--',
         linewidth=5, markersize=15, markeredgecolor='g', markeredgewidth=5, markerfacecolor="r")
plt.title('스타일 적용예')
plt.grid()
plt.show()
```

#### 1.2 그림 범위

```python
plt.title("x축, y축의 범위 설정")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16],
         c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.xlim(0, 50), plt.ylim(-10, 30), plt.grid()
plt.show()
```

- 곡선 그림

```python
X = np.arange(0, 2*np.pi + 0.01, 0.01)
# == X = np.linspace(0, 2*np.pi, 500)
Ysin, Ycos = np.sin(X), np.cos(X)
plt.plot(X, Ysin), plt.plot(X, Ycos)
plt.grid(), plt.title('Sine, Cosine 그래프')
plt.show()
```

#### 1.3 Tick 설정

```python
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
plt.yticks([-1, 0, 1])
plt.title('Tick 설정'),plt.plot(X, Ysin), plt.plot(X, Ycos), plt.grid()
plt.show()
```

```python
X = np.arange(0, 2*np.pi + 0.01, 0.01)
Ysin, Ycos = np.sin(X), np.cos(X)
plt.title("LaTeX, 문자열로 tick label 정의")
plt.plot(X, Ysin), plt.plot(X, Ycos), plt.grid()
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$0$', r'$\pi/2$', r'$+\pi$', r'$+3\pi/2$', r'$+2\pi$'])
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.show()
```

- 범례(legend)

```python
plt.plot(X, Ysin, label='sin'), plt.plot(X, Ycos, label='cos'), plt.grid()
plt.title("범례 표시")
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$0$', r'$\pi/2$', r'$+\pi$', r'$+3\pi/2$', r'$+2\pi$'])
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.legend(loc=9)       # 범례 위치 loc= 0~10 값 혹은 'upper center'꼴의 텍스트로 조절가능
plt.show()
```

- X축, Y축 label

```python
plt.plot(X, Ysin, label='sin'), plt.plot(X, Ycos, label='cos'), plt.grid()
plt.title("X축, Y축 label 표시")
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$0$', r'$\pi/2$', r'$+\pi$', r'$+3\pi/2$', r'$+2\pi$'])
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.legend(loc=0)
plt.xlabel('시간'), plt.ylabel('진폭')
plt.show()
```

- 그림의 크기 지정

```python
plt.figure(figsize=(10, 6))
plt.plot(X, Ysin, label='sin'), plt.plot(X, Ycos, label='cos'), plt.grid()
plt.title("그림 크기 지정")
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$0$', r'$\pi/2$', r'$+\pi$', r'$+3\pi/2$', r'$+2\pi$'])
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.legend(loc=9)       # 범례 위치 loc= 0~10 값 혹은 'upper center'꼴의 텍스트로 조절가능
plt.show()
```

- 그림 저장

```python
plt.figure(figsize=(10, 6))
plt.plot(X, Ysin, label='sin'), plt.plot(X, Ycos, label='cos'), plt.grid()
plt.title("그림 저장", fontsize=15)
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$0$', r'$\pi/2$', r'$+\pi$', r'$+3\pi/2$', r'$+2\pi$'])
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.legend(loc=9)       # 범례 위치 loc= 0~10 값 혹은 'upper center'꼴의 텍스트로 조절가능
plt.savefig('data/삼각함수.png')
```
