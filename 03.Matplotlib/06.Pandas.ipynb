#### 5. Pandas 시각화

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

- Line graph

```python
np.random.seed(2023)
df = pd.DataFrame(np.random.randn(100, 3), columns=list('ABC'),
                  index=pd.date_range('2023-04-01', periods=100)).cumsum()      # cumsum : 누계
df.head()
```

```python
df.plot(figsize=(5,3))           # default가 line graph
plt.title('Pandas의 plot method 사용 예')
plt.xlabel('일자'), plt.ylabel('데이터')
plt.show()
```

- Bar chart

```python
iris = sns.load_dataset('iris')

iris.sepal_length[:30].plot(kind='bar', figsize=(8, 3), rot=360, width=0.5)
plt.title('Iris Setosa 품종의 sepal length');
```

```python
iris.sepal_length[:30].plot(kind='barh', figsize=(8, 5), rot=360, width=0.5);
```

```python
df2 = iris.groupby('species').mean()
df2
```

```python
df2.plot.bar(rot=0)
plt.title('Iris 품종별 Feature의 평균');
```

```python
df2.columns.name = 'Feature'
df2.plot.bar(rot=0)
plt.ylim(0, 8)                 # legend <=> graph 간격 띄우기
plt.title('Iris 품종별 Feature의 평균');
```

- Pie chart

```python
titanic = sns.load_dataset('titanic')

s = titanic['class'].value_counts()
s
```

```python
s.plot.pie(autopct='%.1f%%')
plt.title('선실별 승객 수');
```

- histogram, kernel density

```python
iris.plot.hist()
iris.plot(kind='kde')           # scipy 모듈 설치
plt.show()
```

- box plot

```python
iris[iris.species == 'setosa'].plot.box()
plt.title('Setosa의 각 feature별 Box plot');
```

```python
iris.boxplot(by='species')
plt.tight_layout()
plt.suptitle('Iris 각 피처의 품종별 박스 플롯');
```
