#### 1. 여러개의 그래프

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
np.random.seed(0)
f1 = plt.figure(figsize=(10, 2))
plt.title("figure size : (10, 2)")
plt.plot(np.random.randn(100), color='blue'), plt.grid(color='palegreen')
plt.show()
```

- subplot()

```python
from PIL import Image
img1 = Image.open('data/output.png')
img2 = Image.open('data/삼각함수.png')
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1, 2, 1)
plt.imshow(img1), plt.title('라인 그래프')
plt.subplot(122)
plt.title('삼각함수 그래프'), plt.imshow(img2)
plt.show()
```

```python
ans = sns.load_dataset('anscombe')
x1 = np.linspace(4, 20)
y = 3 + 0.5*x1

plt.figure(figsize=(10,8))
for index, i in enumerate(ans.dataset.unique()):
    ds = ans[ans.dataset == i]
    plt.subplot(2, 2, index+1)
    plt.scatter(ds.x, ds.y, color='blue')
    plt.plot(x1, y, 'r')
    plt.title('Anscombe quartet'), plt.grid(color='green')
plt.show()
```

- subplots() : axis 공유, supertitle설정 등이 필요할때 사용

```python
fig, axes = plt.subplots(2, 2, figsize=(10,8), sharex=True, sharey=True)

for index, i in enumerate(ans.dataset.unique()):
    ds = ans[ans.dataset == i]
    ax = axes[index//2, index%2]     # row index, column index
    ax.scatter(ds.x, ds.y, color='blue')
    ax.plot(x1, y, 'r')
    ax.set_title('Anscombe quartet'), ax.grid(color='green')

fig.suptitle("Anscombe's Quartet", fontsize=15)
plt.tight_layout()
plt.show()
```

- Axis 객체와 축

```python
fig, ax0 = plt.subplots()
ax1 = ax0.twinx()
ax0.set_title("2개의 y축 한 figure에서 사용하기")
ax0.plot([10, 5, 2, 9, 7], 'r-', label="y0")
ax0.set_ylabel("y0")
ax0.grid()
ax1.plot([100, 200, 220, 180, 120], 'g:', label="y1")
ax1.set_ylabel("y1")
ax1.grid()
ax0.set_xlabel("공유되는 x축")
fig.legend()
plt.show()
```

- matplotlib 갤러리에 있는 그래프

```python
import numpy as np
import matplotlib.pyplot as plt


x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.figure(figsize=(10,6))
plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-', color='blue', markerfacecolor='r')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
plt.grid(color='gray')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-', color='blue', markerfacecolor='r')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
plt.grid(color='gray')

plt.show()
```

- Iris 모든 품종의 특성에 대한 Box plot을 하나의 그래프 안에 넣기

```python
iris = sns.load_dataset('iris')

plt.figure(figsize=(10, 7))

for index, species in enumerate(iris.species.unique()):
    sl = iris[iris.species == species].sepal_length
    sw = iris[iris.species == species].sepal_width
    pl = iris[iris.species == species].petal_length
    pw = iris[iris.species == species].petal_width
    plt.subplot(1, 3, index+1)
    plt.boxplot((sl, sw, pl, pw), labels=['sl', 'sw', 'pl', 'pw'])
    plt.title(species), plt.grid()
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(10,8), sharex=True, sharey=True)

for index, species in enumerate(iris.species.unique()):
    sl = iris[iris.species == species].sepal_length
    sw = iris[iris.species == species].sepal_width
    pl = iris[iris.species == species].petal_length
    pw = iris[iris.species == species].petal_width
    ax = axes[index]     # row index, column index
    ax.boxplot((sl, sw, pl, pw), labels=['sl', 'sw', 'pl', 'pw'])
    ax.set_title(species), ax.grid(color='green')

fig.suptitle("flower", fontsize=15)
plt.tight_layout()
plt.show()
```
