### 경기도 공원
    - 시군구별 공원면적 -> 단계구분도
    - 시군구별 인당 공원면적 -> 단계구분도

```python
import requests
from urllib.parse import quote
import pandas as pd
import folium, json
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('../data/경기도공원0421.csv', encoding='euc-kr')
df.소재지도로명주소[437] = '경기도 수원시 팔달구 덕영대로 735번길 18'
```

```python
# 필요한 컬럼 filtering
df = df['공원명 소재지도로명주소 위도 경도 공원면적'.split(' ')]
df
```

```python
# 자치구가 있는 도시 데이터 전처리
gu_city = ['수원시', '고양시', '용인시', '성남시', '안산시', '안양시']
for index in df.index:
    if df.소재지도로명주소[index].split()[1] in gu_city:
        df.소재지도로명주소[index] = ''.join(df.소재지도로명주소[index].split()[1:3])
df
```

```python
# 자치구가 없는 도시 데이터

for index in df.index:
    df.소재지도로명주소[index] = df.소재지도로명주소[index].split()

for index in df.index:
    if len(df.소재지도로명주소[index]) > 1:
        df.소재지도로명주소[index] = df.소재지도로명주소[index][1]
    else:
        df.소재지도로명주소[index] = df.소재지도로명주소[index][0]
df.소재지도로명주소
```

```python
city_parks = df.groupby('소재지도로명주소')['공원면적'].sum()
city_parks
```

```python
pop = pd.read_csv('../data/경기도인구소멸비율.csv')
pop
```

```python
temp = pop.groupby('시군구')['총 인구수'].sum()
df2 = pd.DataFrame(temp)
df2
```

```python
df1 = pd.DataFrame(city_parks)
df1
```

```python
pops = []
for index1 in df1.index:
    for index2 in df2.index:
        if index1 == index2:
            pops.append(df2['총 인구수'][index2])
pops
```

```python
df1['총인구수'] = pops
```

```python
df1.to_csv('../data/경기도공원인구', index=False)
```

```python
file_name = '../data/경기도_geo_simple.json'
geo_data = json.load(open(file_name, encoding='utf-8'))
```

```python
map = folium.Map([37.4, 127.1], zoom_start=9, tiles='stamen toner')
folium.Choropleth(
    geo_data=geo_data,
    data=df1.공원면적,                       # 단계구분도로 보여줄 데이터
    columns=[df1.index, df1.공원면적],        # DF에서 추출할 컬럼
    fill_color='Pastel2',                      # colormap
    key_on='properties.name'
).add_to(map)

map
```

### 서울시 공원
    - 자치구별 공원면적 -> 단계구분도
    - 자치구별 인당 공원면적 -> 단계구분도

```python
pp = pd.read_csv('../data/서울시인구.txt', sep='\t')
pp
```

```python
for i in range(len(pp.columns)):
    for j in pp.index:
        pp.iloc[j][i] = pp.iloc[j][i].replace(',', '')
pp.head()
```

```python
pp.drop(index=([0, 1]), inplace=True)
```

```python
pp = pp[['자치구', '인구']]
pp.drop(index=2, inplace=True)
pp
```

```python
parks = pd.read_csv('../data/서울시공원현황.csv', encoding='euc-kr')
parks = parks[['공원명', '면적', '공원주소', 'X좌표(WGS84)', 'Y좌표(WGS84)']]
parks.dropna(subset=['면적', 'X좌표(WGS84)'],inplace=True)
parks.head(4)
```

```python
# 이상한 데이터 삭제
parks.drop([2], inplace=True)
parks.drop([6], inplace=True)
```

```python
parks.면적 = parks.면적.str.extract(r'(\d+)').astype(int)
parks
```

```python
parks.공원주소 = parks.공원주소.str.split().apply(lambda x: x[1])
```
