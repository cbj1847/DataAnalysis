### 경기도 공원
    - 시군구별 공원면적 -> 단계구분도
    - 시군구별 인당 공원면적 -> 단계구분도

```python
import requests
from urllib.parse import quote
import pandas as pd
import folium, json
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('../data/전국도시공원정보표준데이터.csv', encoding='euc-kr')
# df.소재지도로명주소[437] = '경기도 수원시 팔달구 덕영대로 735번길 18'
df
```

```python
for index in df.index:
    df= df[df.소재지도로명주소[index].split()[0] == '경기도']
df
```

```python
# 필요한 컬럼 filtering
df = df['공원명 소재지도로명주소 위도 경도 공원면적'.split(' ')]
df
```

```python
# 자치구가 있는 도시 데이터 전처리
gu_city = ['수원시', '고양시', '용인시', '성남시', '안산시', '안양시']
for index in df.index:
    if df.소재지도로명주소[index].split()[1] in gu_city:
        df.소재지도로명주소[index] = (df.소재지도로명주소[index].split()[1:3])
df
```

```python
# 자치구가 없는 도시 데이터

for index in df.index:
    df.소재지도로명주소[index] = df.소재지도로명주소[index].split()

for index in df.index:
    if len(df.소재지도로명주소[index]) > 1:
        df.소재지도로명주소[index] = df.소재지도로명주소[index][1]
    else:
        df.소재지도로명주소[index] = df.소재지도로명주소[index][0]
print(df.소재지도로명주소)
```

```python
city_parks = df.groupby('소재지도로명주소')['공원면적'].sum()
city_parks
```

```python
pop2 = pd.read_csv('../data/주민등록인구집계현황.csv', encoding='euc-kr')
pop2 = pop2[pop2.연도 == 2023]
pop2 = pop2[pop2.월 == 3]
pop2 = pop2[['행정구역명', '총 인구수']]
pop2
```

```python
for index in pop2.index:
    if pop2['총 인구수'][index] == 0:
        pop2.drop(index=index, inplace=True)
pop2.drop(index=0, inplace=True)
pop2
```

```python
# 자치구가 있는 도시 데이터 전처리
gu_city = ['수원시', '고양시', '용인시', '성남시', '안산시', '안양시']
for index in pop2.index:
    if pop2.행정구역명[index].split()[1] in gu_city:
        pop2.행정구역명[index] = ''.join(pop2.행정구역명[index].split()[1:3])
```

```python
# 자치구가 없는 도시 데이터
for index in pop2.index:
    pop2.행정구역명[index] = pop2.행정구역명[index].split()

for index in pop2.index:
    if len(pop2.행정구역명[index]) > 1:
        pop2.행정구역명[index] = pop2.행정구역명[index][1]
    else:
        pop2.행정구역명[index] = pop2.행정구역명[index][0]
print(pop2.행정구역명)
```

```python
temp = pop2.groupby('행정구역명')['총 인구수'].sum()
df2 = pd.DataFrame(temp)
df2
```

```python
df1 = pd.DataFrame(city_parks)
df1
```

```python
pops = []
for index1 in df1.index:
    for index2 in df2.index:
        if index1 == index2:
            pops.append(df2['총 인구수'][index2])
pops
```

```python
df1['총인구수'] = pops
df1
```

```python
df1['인구당공원면적'] = df1.공원면적 / df1.총인구수
df1
```

```python
df1.to_csv('../data/경기도공원인구', index=False)
```

### 서울시 공원
    - 자치구별 공원면적 -> 단계구분도
    - 자치구별 인당 공원면적 -> 단계구분도

```python
pp = pd.read_csv('../data/서울시인구.txt', sep='\t')
pp
```

```python
for i in range(len(pp.columns)):
    for j in pp.index:
        pp.iloc[j][i] = pp.iloc[j][i].replace(',', '')
pp.head()
```

```python
pp.drop(index=([0, 1]), inplace=True)
```

```python
pp = pp[['자치구', '인구']]
pp.drop(index=2, inplace=True)
```

```python
pp.set_index('자치구', inplace=True)
pp
```

```python
parks = pd.read_csv('../data/서울시공원현황.csv', encoding='euc-kr')
parks = parks[['공원명', '면적', '공원주소', 'X좌표(WGS84)', 'Y좌표(WGS84)']]
parks.dropna(subset=['면적', 'X좌표(WGS84)'],inplace=True)
parks.공원주소[129] = '서울특별시 노원구 공릉2동 산 82-2'
parks.head(4)
```

```python
# 이상한 데이터 삭제
parks.drop([2], inplace=True)
parks.drop([6], inplace=True)
```

```python
parks.면적 = parks.면적.str.extract(r'(\d+)').astype(int)
parks
```

```python
parks.공원주소 = parks.공원주소.str.split().apply(lambda x: x[1])
```

```python
seoul_parks = parks.groupby('공원주소')['면적'].sum()
df3 = pd.DataFrame(seoul_parks)
df3
```

```python
pops2 = []
for index1 in df3.index:
    for index2 in pp.index:
        if index1 == index2:
            pops2.append(int(pp['인구'][index2]))
pops2
```

```python
df3['인구'] = pops2
df3
```

```python
df3['인구당공원면적'] = df3['면적'] / df3['인구']
df3.to_csv('../data/서울시공원인구', index=False)
df3.head(3)
```
