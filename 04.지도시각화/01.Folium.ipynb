### Folium 지도 시각화

- 설치 : pip install folium

```python
import folium
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
folium.__version__
```

```python
map = folium.Map(location=[37.295, 127.045], zoom_start=15)
folium.Marker(
    location=[37.294837, 127.0453902],          # WGS84 좌표계, (latitude,longtitude)
    popup=folium.Popup('경기도 수원시 영통구 도청로 30', max_width=200),
    tooltip='경기도청'
).add_to(map)
map
```

```python
map = folium.Map(location=[37.295, 127.045], zoom_start=15, tiles='Stamen Toner')       # defalult : tiles='Open street map'
folium.Marker(
    location=[37.294837, 127.0453902],          # WGS84 좌표계, (latitude,longtitude)
    popup=folium.Popup('경기도 수원시 영통구 도청로 30', max_width=200),
    tooltip='경기도청'
).add_to(map)
map
```

- 경기도청, 장안구청, 팔달구청, 권선구청, 영통구청

```python
map = folium.Map(location=[37.28, 127.02], zoom_start=12)       # defalult : tiles='Open street map'
folium.Marker(
    location=[37.294837, 127.0453902],          # WGS84 좌표계, (latitude,longtitude)
    popup=folium.Popup('경기도 수원시 영통구 도청로 30', max_width=200),
    tooltip='경기도청'
).add_to(map)
folium.Circle(
    radius=300, color='crimson', fill=True,
    location=[37.30337225, 127.0096118],          # WGS84 좌표계, (latitude,longtitude)
    popup=folium.Popup('경기도 수원시 장안구 송원로 101', max_width=200),
    tooltip='장안구청'
).add_to(map)
folium.CircleMarker(
    radius=15, color='#3186cc', fill=True, fill_color='#3186cc',
    location=[37.28256171, 127.0190196],          # WGS84 좌표계, (latitude,longtitude)
    popup=folium.Popup('경기도 수원시 팔달구 창룡대로 23', max_width=200),
    icon=folium.Icon('blue', icon='star'),
    tooltip='팔달구청'
).add_to(map)
folium.Marker(
    location=[37.25764489,126.9719386],          # WGS84 좌표계, (latitude,longtitude)
    popup=folium.Popup('경기도 수원시 권선구 호매실로 12', max_width=200),
    icon=folium.Icon('purple', icon='glyphicon glyphicon-thumbs-up', prefix='glyphicon'),
    tooltip='권선구청'
).add_to(map)
folium.Marker(
    location=[37.2596127, 127.0465469],          # WGS84 좌표계, (latitude,longtitude)
    popup=folium.Popup('경기도 수원시 영통구 효원로 407', max_width=200),
    icon=folium.Icon('blue', icon='cloud'),
    tooltip='영통구청'
).add_to(map)
title_html = "<h3 align='center' style='font-size:20px'>수원 소재 주요 관공서</h3>"
map.get_root().html.add_child(folium.Element(title_html))
map
```
