### 수원소재 관공서 지도위에 시각화

```python
import folium
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/경기도주요관공서(좌표)')
df['color'] = ['red', 'green', 'blue', 'orange', 'black']
df
```

```python
# for문으로 5개 수원시내 구청 위치 표시
map = folium.Map(location=[df['위도'].mean(), df['경도'].mean()], zoom_start=13)

for i in df.index:
    folium.Circle(
    location=[df.위도[i],df.경도[i]],          # WGS84 좌표계, (latitude,longtitude)
    popup=folium.Popup(df.주소[i], max_width=200),
    radius=300, color=df.color[i], fill=True,
    tooltip=df.이름[i]
    ).add_to(map)

title_html = "<h3 align='center' style='font-size:20px'>수원 소재 주요 관공서</h3>"
map.get_root().html.add_child(folium.Element(title_html))
map
```
