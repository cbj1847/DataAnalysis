### 지식IN - 어버이날 선물 서칭결과 말뭉치(corpus) 말들기

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import warnings
import math
warnings.filterwarnings('ignore')
```

```python
text_list = []
driver = webdriver.Chrome('C:/Users/YONSAI/Desktop/chromedriver_win32/chromedriver.exe')
for i in range(1,11):
    url = f'https://kin.naver.com/search/list.naver?query=%EC%96%B4%EB%B2%84%EC%9D%B4%EB%82%A0%20%EC%84%A0%EB%AC%BC&page={i}'
    driver.get(url)
    time.sleep(2)
    lis = driver.find_elements(By.XPATH, '//*[@id="s_content"]/div[3]/ul/li')
    for li in lis:
        driver = webdriver.Chrome('C:/Users/YONSAI/Desktop/chromedriver_win32/chromedriver.exe')
        driver.get(url)
        lis = driver.find_elements(By.XPATH, '//*[@id="s_content"]/div[3]/ul/li')  
        href = li.find_element(By.CSS_SELECTOR, 'dt > a').get_attribute('href')
        driver.get(href)
        time.sleep(2)
        if len(driver.find_element(By.CLASS_NAME, '_nextPageAnswerCount').text) != 0:
            while(len(driver.find_element(By.CLASS_NAME, '_nextPageAnswerCount').text) != 0):
                driver.find_element(By.ID, 'nextPageButton').click()
                time.sleep(2)
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        texts = soup.select('.se-module.se-module-text')
        for text in texts:
            text_list.append(text.get_text())
driver.close()
```
