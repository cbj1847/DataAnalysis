### 인구소멸 위기지역 Top10

- 기초자치단체
    - 특례시 : 수원, 고양, 용인
    - 행정구 : 성남, 안산, 안양
    - 가임기(20-39)여성 / 60세이상 노령인구

```python
import numpy as np
import pandas as pd
```

```python
# read file
df1 = pd.read_csv('../data/주민등록인구집계현황.csv', encoding='euc-kr')
df2 = pd.concat([df1.iloc[:, :4], df1.iloc[:, -12:]] ,axis=1)
df2.drop(index=0, axis=1, inplace=True)
df2.isna().sum()

# 2022년 12월 자료 필터링
df2 = df2[df2['연도'] == 2022]
df2 = df2[df2['월'] == 12]

# 가임기여성, 노령여성인구 열 추가
df2['가임여성인구'] = df2.iloc[:, 5] + df2.iloc[:, 6]
df2['노령여성인구'] = df2.iloc[:, -5:].sum(axis=1)
df2 = pd.concat([df2.iloc[:, 0:4], df2.iloc[:, -2:]], axis=1)
df2['가임여성비율'] = (df2.iloc[:, -2] / df2.iloc[:, -1] * 100).round(2)

#50% 미만 지역 필터링
df3 = df2['가임여성비율'] <= 50
df2 = df2[df3]

#정렬
df2.sort_values(by='가임여성비율', inplace=True)
df2
```
