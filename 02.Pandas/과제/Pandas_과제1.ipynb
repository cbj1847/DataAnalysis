1. cctv 데이터를 다음과 같이 만들어 cctv.csv 파일에 저장하세요.

```python
df = pd.read_csv('../data/서울시 자치구 년도별 CCTV 설치 현황_221231기준.csv', encoding='euc-kr', thousands=',')
df
```

```python
import warnings
warnings.filterwarnings('ignore')

df1 = df[1:]

df1 = df1.replace(',', '')
df2 = df1['구분']
df3 = df1['총계']
df_3years = pd.DataFrame([], columns=['구별', 'CCTV댓수', '최근증가율'])
df_3years['구별'] = df2
df_3years['CCTV댓수'] = df3

rate = pd.Series()
rate3 = pd.Series()

for i in range(0,25):
    temp = 0
    for j in range(1,4):
        temp += df1.iloc[i][-j]
    rate[i] = df1.iloc[i][1] - temp
    rate3[i] = temp

result = rate3 / rate * 100

for i in range(25): 
    df_3years['최근증가율'][i+1] = result[i].round(2)

df_3years
```

2. 서울시 인구 데이터를 다음과 같이 만들어 pop.csv 파일에 저장하세요.

```python
pp = pd.read_csv('../data/서울시인구.txt', sep='\t', thousands=',')
pp.head()
```

```python
for i in range(len(pp.columns)):
    for j in pp.index:
        pp.iloc[j][i] = pp.iloc[j][i].replace(',', '')
pp.head()
```

```python
pp1 = pd.DataFrame([], columns=['구별', '인구수', '내국인', '외국인', '고령자', '외국인비율', '고령자비율'])

temp = []
for i in pp.index[3:]:
    temp.append(pp.iloc[i][1])
pp1['구별'] = temp

temp = []
for i in pp.index[3:]:
    temp.append(int(pp.iloc[i][3]))
pp1['인구수'] = temp

temp = []
for i in pp.index[3:]:
    temp.append(int(pp.iloc[i][6]))
pp1['내국인'] = temp

temp = []
for i in pp.index[3:]:
    temp.append(int(pp.iloc[i][9]))
pp1['외국인'] = temp

temp = []
for i in pp.index[3:]:
    temp.append(int(pp.iloc[i][13]))
pp1['고령자'] = temp

pp1['외국인비율'] = (pp1['외국인'] / pp1['인구수'] * 100).round(2)

pp1['고령자비율'] = (pp1['고령자'] / pp1['인구수'] * 100).round(2)

pp1
```
