- tips data

```python
print(tips.shape)
tips.head()
```

```python
tips['tip_pct'] = (tips.tip / tips.total_bill * 100).round(2)
tips.tail()
```

```python
tips.describe()
```

```python
# 성별, 흡연유무별 팁의 비율
tips.groupby(['sex', 'smoker'])[['tip_pct']].mean().round(3)
```

```python
# 성별, 흡연유무별 팁의 비율의 분석값들
tips.groupby(['sex', 'smoker'])[['tip_pct']].describe().round(3)
```

```python
# 요일, 시간별 팁의 비율의 분석값들
tips.groupby(['day', 'time'])[['tip_pct']].describe().round(3)
```

##### 2.2 pivot_table

```python
# 성별, 흡연유무별 팁의 비율
tips.pivot_table('tip_pct', 'sex', 'smoker', ['std', 'mean'])       # (values, index, columns, aggfunc)
```

```python
# 합계 기능
# 성별, 흡연유무별 인원
tips.pivot_table('tip_pct', 'sex', 'smoker', 'count', margins=True, margins_name='계')
```

##### 2.3 groupby vs pivot_table

```python
# 성별 팁 비율의 평균
tips.groupby('sex')[['tip_pct']].mean()
```

```python
tips.pivot_table('tip_pct', 'sex')
```

```python
# 흡연 유무에 따른 평균 팁 비율과 최소 / 최대 값
tips.groupby('smoker')[['tip_pct']].agg(['mean', 'min', 'max'])
```

```python
tips.pivot_table('tip_pct', 'smoker', aggfunc=['mean', 'min', 'max'])
```

- 연습문제 4.7.4
    - 타이타닉 승객 데이터를 이용하여 다음 분석을 실시하라. 데이터는 다음과 같이 받을 수 있다.
    - titanic = sns.load_dataset("titanic")
    - qcut 명령으로 세 개의 나이 그룹을 만든다.

    - 성별, 선실, 나이 그룹에 의한 생존율을 데이터프레임으로 계산한다. 행에는 성별 및 나이 그룹에 대한 다중 인덱스를 사용하고 열에는 선실 인덱스를 사용한다. - 생존률은 해당 그룹의 생존 인원수를 전체 인원수로 나눈 값이다.

    - 성별 및 선실에 의한 생존율을 피봇 데이터 형태로 만든다.

```python
df = sns.load_dataset('titanic')
df.head(3)
```

```python
labels1 = ['청년', '중년', '장년']
df['나이그룹'] = pd.qcut(df['age'], 3, labels=labels1)
df
```

```python
df.groupby(['sex', 'pclass', 'age'])['survived']
```

```python
# 성별 및 선실에 의한 생존율
df.pivot_table('survived', 'sex', 'class')
```
